```
%time 

#lien data gouv https://www.data.gouv.fr/en/datasets/demandes-de-valeurs-foncieres/
#faq https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20191030-171734/faq.pdf
#dictionnaire de donnÃ©es https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20191220-102114/notice-descriptive-du-fichier-dvf.pdf
import pandas as pd
from tqdm.notebook import tqdm

url_dict = {
    '2019':'https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20200416-115822/valeursfoncieres-2019.txt',
    '2018':'https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20200416-115643/valeursfoncieres-2018.txt',
    '2017':'https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20200416-115446/valeursfoncieres-2017.txt',
    '2016':'https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20200416-115256/valeursfoncieres-2016.txt',
    '2015':'https://static.data.gouv.fr/resources/demandes-de-valeurs-foncieres/20200416-114842/valeursfoncieres-2015.txt'
}
def get_subsample():
    data_all_year=[]
    for year,url in tqdm(url_dict.items()):
        print(f'----------STARTING YEAR {year}----------')
        data_ = pd.read_csv(url, sep="|")
        data_ = data_.drop(['Code service CH','Reference document','1 Articles CGI','2 Articles CGI',
                                      '3 Articles CGI','4 Articles CGI','5 Articles CGI','Identifiant local','No Volume','B/T/Q',
                                      'Surface Carrez du 1er lot','2eme lot','Surface Carrez du 2eme lot','3eme lot',
                                      'Surface Carrez du 3eme lot','4eme lot','Surface Carrez du 4eme lot','5eme lot',
                                      'Surface Carrez du 5eme lot','Nature culture','Nature culture speciale','Surface terrain'], axis=1)
        
        print('Shape from import: ',data_.shape)

        marseille_list_arrondissement = ['MARSEILLE 8EME','MARSEILLE 9EME','MARSEILLE 13EME','MARSEILLE 5EME',
                                         'MARSEILLE 10EME','MARSEILLE 12EME','MARSEILLE 6EME','MARSEILLE 11EME',
                                         'MARSEILLE 4EME','MARSEILLE 15EME','MARSEILLE 1ER','MARSEILLE 3EME',
                                         'MARSEILLE 7EME','MARSEILLE 2EME','MARSEILLE 14EME','MARSEILLE 16EME']

        data_marseille = data_[data_['Commune'].isin(marseille_list_arrondissement)]
        print('Shape after subsample: ',data_marseille.shape)
        print(f'\n Sampling of {round((data_marseille.shape[0]/data_.shape[0])*100,2)} % of the original dataframe \n')
        del data_
        data_marseille['year'] = year
        data_all_year.append(data_marseille)
        del data_marseille
        print(f'----------YEAR {year} DONE---------- \n')
        
    print('STARTING CONCATENATING')
    df = pd.concat(data_all_year)
    print('CONCATENING DONE')
    return df

df = get_subsample()
```
